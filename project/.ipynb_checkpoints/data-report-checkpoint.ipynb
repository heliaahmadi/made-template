## Introduction
This report analyzes the variation in energy consumption and CO2 emissions across London boroughs using data from the London Datastore. This project aims to identify patterns and trends that could inform future environmental and urban planning policies.

```python
import pandas as pd
import sys
import io
import requests
from sqlalchemy import create_engine
import os
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
def load_data(url):
    try:
        response = requests.get(url)
        df = pd.read_csv(io.StringIO(response.content.decode('utf-8')))
        return df
    except Exception as e:
        print(f"Error loading data: {e}")
        return None

def clean_data(df):
    try:
        df_cleaned = df.dropna()
        return df_cleaned
    except Exception as e:
        print(f"Error cleaning data: {e}")
        return None
```

## Data Sources
- **Data Description**: The datasets include annual measures of energy consumption and CO2 emissions per borough in London.
- **Source**: London Datastore (https://data.london.gov.uk/)
- **License**: Data is available under the UK Open Government Licence, permitting free use and redistribution.


```python
# Cell 5: Code
urls = [
    "https://data.london.gov.uk/download/leggi/2d6ee3f1-e928-48a9-8eab-01748c65ac6f/energy-consumption-borough-leggi.csv",
    "https://data.london.gov.uk/download/leggi/fcf8a0a3-2051-484a-ba9a-5c8bc2268a3e/co2-emissions-borough-leggi.csv"
]
names = ["energy_consumption", "co2_emissions"]

# Load and clean data
data_frames = {}
for name, url in zip(names, urls):
    df = load_data(url)
    df_cleaned = clean_data(df)
    data_frames[name] = df_cleaned
```

```python
import matplotlib.ticker as ticker

co2_data = data_frames['co2_emissions']
selected_boroughs = ['Camden', 'Westminster', 'Kensington and Chelsea', 'Greenwich', 'City of London']
filtered_data = co2_data[co2_data['Borough'].isin(selected_boroughs)]

print(filtered_data)

plt.figure(figsize=(12, 8))
ax = sns.lineplot(data=filtered_data, x='LEGGI_Year', y='KtCO2e', hue='Borough', marker='o')
plt.title('CO2 Emissions Over Time by Selected Boroughs')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (Metric Tons)')
plt.legend(title='Borough', loc='upper right', labels=selected_boroughs)
plt.grid(True)
ax.yaxis.set_major_locator(ticker.MultipleLocator(50))  # Adjust the 50 value as needed for your scale
ax.yaxis.set_major_formatter(ticker.ScalarFormatter())
plt.show()






```
